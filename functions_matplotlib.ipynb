```python
# Ejercicio 1 A
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10,10, 1000)
f = x.astype(int)
f = np.cbrt(x) + np.floor(x/2)
plt.plot(x,f, '.',color="coral")

plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid()

plt.title ( 'Gráfica de la funcion f(x) = x^(1/3) + [x/2]' )

""" 
f(x) = x**(1/3) + [x/2] = ∞
lim x --> ∞   

La funcion no tiene puntos de discontinuidad 
La funcion no tiene AV ni AH.
"""
```

```python
# Ejercicio 3 
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(-5,5,1000)
f= (x-1)* abs(x-1)
t1= 2*x - 1
fig= plt.figure(figsize= (8,10))

plt.plot(x,f)
plt.plot(x,t1,c="red")
plt.xlabel( "Eje X")
plt.ylabel("Eje Y")
plt.grid()
plt.title("Gráfica de las funciones f(x)= (x-1) * |x-1|     y = 2*x-1 ")
plt.legend(['F(x)= (x-1) * |x-1|', 'Recta tangente: Y = 2*x - 1', ] , title='Referencia:', loc= 'upper center')
```
